```python
from sklearn.metrics import cohen_kappa_score
import pandas as pd
import numpy as np
import pylab as pl


# Variable of annex levels
annex1Level = ''
annex2Level = ''

# Variable of annex class
annexClass = ''

# List of files
list_files = []

# Get name files
anexo1 = ("anexo.csv")
anexo2 = ("anexo-grex.csv")
list_files.append(anexo1)
list_files.append(anexo2)
print(list_files)

# Read csv files
for file in list_files:
        index = list_files.index(file)
        reader = pd.read_csv(file)
        if index == 0:
            annex1Level = (reader.iloc[1:, 2].to_list())
            annexClass = (reader.iloc[1:, 1].to_list())
        else:
            annex2Level = (reader.iloc[1:, 2].to_list())
print('\nLEVELS: ')      
print('\nFirst version:  ' + str(annex1Level))
print('\nSecond version: ' + str(annex2Level))


# Save class and level of each annex in result
result = ''
for i in range(0, len(annexClass)):
    result += ('Ana:  ' + annexClass[i] + ': ' + annex1Level[i])
    result += ('\nGrex: ' + annexClass[i] + ': ' + annex2Level[i] + '\n\n')

# Write result in file.txt
file = open("comparative_classes.txt", "w")
file.write(result)
file.close()

    
# List of annex
list_annex = [annex1Level, annex2Level]


# Assign value and count the number of elements at each level.
levels = ''
for i in range(0, len(list_annex)):
    level_A1, level_A2, level_B1, level_B2, level_C1, level_C2 = 0,0,0,0,0,0
    if i == 0:
        levels += ('Ana:  ')
    else:
        levels += ('\n\nGrex: ')
    for value in range(0, len(list_annex[i])):
        #print(list_annex[i][value])
        if (list_annex[i][value] == 'A1'):
            level_A1 += 1
            list_annex[i][value] = 1
        elif (list_annex[i][value] == 'A2'):
            level_A2 += 1
            list_annex[i][value] = 2
        elif (list_annex[i][value] == 'B1'):
            level_B1 += 1
            list_annex[i][value] = 3
        elif (list_annex[i][value] == 'B2'):
            level_B2 += 1
            list_annex[i][value] = 4
        elif (list_annex[i][value] == 'C1'):
            level_C1 += 1
            list_annex[i][value] = 5
        elif (list_annex[i][value] == 'C2'):
            level_C2 += 1
            list_annex[i][value] = 6
        else:
            list_annex[i][value] = 0
    levels += ('\nElements of A1 level: ' + str(level_A1))
    levels += ('\nElements of A2 level: ' + str(level_A2))
    levels += ('\nElements of B1 level: ' + str(level_B1))
    levels += ('\nElements of B2 level: ' + str(level_B2))
    levels += ('\nElements of C1 level: ' + str(level_C1))
    levels += ('\nElements of C2 level: ' + str(level_C2))

# Write levels in file.txt
file = open("comparative_levels.txt", "w")
file.write(levels)
file.close()

# Assign the values
print('\nASSIGNMENTS: ')
print('\nFirst version:  ' + str(annex1Level))
print('\nSecond version: ' + str(annex2Level))

```

```python
# Calculating the kappa coefficient total
kappa = cohen_kappa_score(annex1Level, annex2Level)
print('\nKappa coefficient: ' + str(kappa))
      
# Draw graph
X = np.linspace(0, 83, 83, endpoint=True)
pl.plot(X, annex1Level, color="blue", linewidth=1.0, linestyle="-", label= "ana")
pl.plot(X, annex2Level, color="green", linewidth=1.0, linestyle="-", label = "grex")

# Show legend
pl.legend(loc='lower right')

pl.title("Comparative level graph")  # Sets the title of the graph
pl.xlabel("Elements")  # Sets the title of the x-axis
pl.ylabel("Levels")   # Sets the title of the y-axis

pl.savefig("Anexo-kappa1.jpg")

# Show result on screen
pl.show()
```

```python
# Get index of elements A1-B1
annex1Level_B1 = []
annex2Level_B1 = []

for i in annex1Level:
    #print(i)
    if ((i == 1) or (i == 2) or (i == 3)):
        index_elem = annex1Level.index(i)
        annex1Level_B1.append(annex1Level[index_elem])
        annex2Level_B1.append(annex2Level[index_elem])
print('\nFirst version: ' + str(annex1Level_B1))
print('\nSecond version: ' + str(annex2Level_B1))


# Calculating length of list
length = len(annex1Level_B1)

# Calculating the kappa coefficient up to 'B1' levels
kappa = cohen_kappa_score(annex1Level_B1, annex2Level_B1)
print('\nKappa coefficient: ' + str(kappa))

# Draw graph
X = np.linspace(0, length, length, endpoint=True)
pl.plot(X, annex1Level_B1, color="blue", linewidth=1.0, linestyle="-", label= "ana")
pl.plot(X, annex2Level_B1, color="green", linewidth=1.0, linestyle="-", label = "grex")

# Show legend
pl.legend(loc='lower right')

pl.title("Comparative level graph up to B1 level")  # Sets the title of the graph
pl.xlabel("Elements")  # Sets the title of the x-axis
pl.ylabel("Levels")   # Sets the title of the y-axis

pl.savefig("Anexo-kappa-B1.jpg")

# Show result on screen
pl.show()
```
